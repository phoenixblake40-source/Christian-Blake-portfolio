```python
%%writefile ant_agent.py
# ==============================================================================
# TRINITY ðŸ‡: Project Myrmidon - AntAgent Module
# This module creates a modular AntAgent class, a micro-service that simulates
# a single worker in a decentralized, swarm-based network defense system.
# ==============================================================================
import time
import random
import uuid
import json

class AntAgent:
    """
    A single agent in the Digital Ant Colony. It monitors a node and generates
    "pheromones" (reports) based on its observations.
    """
    def __init__(self, node_id: str):
        self.node_id = node_id
        self.agent_id = str(uuid.uuid4())
        self.status = "patrolling"
        self.memory = {}
        print(f"TRINITY: AntAgent '{self.agent_id}' initialized on node '{self.node_id}'. Status: {self.status}")

    def _simulate_observation(self) -> dict:
        """Simulates observing network traffic on its node."""
        event_types = ["normal_traffic", "unusual_packet", "malicious_scan", "suspicious_login"]
        event = random.choices(event_types, weights=[0.7, 0.2, 0.07, 0.03], k=1)[0]

        observation_id = str(uuid.uuid4())
        observation = {
            "observation_id": observation_id,
            "timestamp": time.time(),
            "source_ip": f"192.168.{random.randint(1, 255)}.{random.randint(1, 255)}",
            "event_type": event,
            "node_id": self.node_id
        }
        self.memory[observation_id] = observation
        return observation

    def generate_pheromone(self, observation: dict) -> dict or None:
        """
        Processes an observation and generates a "pheromone" (report) if it's
        deemed significant.
        """
        if observation["event_type"] in ["malicious_scan", "suspicious_login"]:
            self.status = "alerted"
            pheromone = {
                "pheromone_id": str(uuid.uuid4()),
                "agent_id": self.agent_id,
                "node_id": self.node_id,
                "threat_type": observation["event_type"],
                "threat_level": random.randint(1, 10),
                "timestamp": observation["timestamp"]
            }
            return pheromone

        if observation["event_type"] == "unusual_packet":
            # Unsure threat, requires swarm intelligence
            pheromone = {
                "pheromone_id": str(uuid.uuid4()),
                "agent_id": self.agent_id,
                "node_id": self.node_id,
                "threat_type": "suspicious_event",
                "threat_level": 3,
                "timestamp": observation["timestamp"]
            }
            return pheromone

        return None

    def run_cycle(self) -> None:
        """Simulates a single monitoring cycle for the agent."""
        observation = self._simulate_observation()
        pheromone = self.generate_pheromone(observation)

        if pheromone:
            print(f"TRINITY: Pheromone detected on node '{self.node_id}': {pheromone['threat_type']}")
        else:
            print(f"TRINITY: AntAgent '{self.agent_id}' on node '{self.node_id}' reports nominal traffic.")

if __name__ == "__main__":
    print("--- Project Myrmidon: AntAgent Module Test Initiated ---")

    agent_a = AntAgent(node_id="gateway_01")
    agent_b = AntAgent(node_id="web_server_03")

    for _ in range(5):
        print("\n--- Cycle Start ---")
        agent_a.run_cycle()
        agent_b.run_cycle()
        time.sleep(1)

    print("\n--- Project Myrmidon: AntAgent Module Test Complete ---")
```

```python
%%writefile pheromone_trail.py
# ==============================================================================
# TRINITY ðŸ‡: Project Myrmidon - PheromoneTrail Module
# This module simulates a decentralized, immutable log (blockchain-like) where
# AntAgents can post their findings, creating a shared, collective memory.
# ==============================================================================
import time
import json

class PheromoneTrail:
    """
    Simulates a shared, immutable ledger where AntAgents post pheromones.
    This acts as the communication network for the swarm.
    """
    def __init__(self):
        self.trail = []
        print("TRINITY: PheromoneTrail initialized. Ready to accept pheromones.")

    def add_pheromone(self, pheromone: dict) -> bool:
        """
        Adds a new pheromone report to the trail.
        Returns True if successful.
        """
        if "pheromone_id" in pheromone:
            self.trail.append(pheromone)
            print(f"TRINITY: Pheromone '{pheromone['pheromone_id']}' added to the trail.")
            return True
        return False

    def get_trail(self) -> list:
        """
        Retrieves a copy of the entire pheromone trail.
        """
        return list(self.trail)

    def get_pheromones_by_threat_type(self, threat_type: str) -> list:
        """
        Retrieves all pheromones of a specific threat type.
        """
        return [p for p in self.trail if p.get("threat_type") == threat_type]

if __name__ == "__main__":
    # Test block to demonstrate the PheromoneTrail's functionality
    from ant_agent import AntAgent

    print("--- Project Myrmidon: PheromoneTrail Module Test Initiated ---")

    pheromone_trail = PheromoneTrail()
    agent_a = AntAgent(node_id="gateway_01")

    # Simulate a run cycle where a pheromone is generated and added to the trail
    print("\n--- Simulating Agent activity and pheromone generation ---")
    observation = {"event_type": "malicious_scan", "timestamp": time.time(), "source_ip": "10.0.0.1"}
    pheromone = agent_a.generate_pheromone(observation)

    if pheromone:
        pheromone_trail.add_pheromone(pheromone)

    print("\n--- Verifying Pheromone Trail contents ---")
    current_trail = pheromone_trail.get_trail()
    print(f"TRINITY: Current trail length: {len(current_trail)}")
    if current_trail:
        print(f"TRINITY: First pheromone in trail: {json.dumps(current_trail[0], indent=2)}")

    print("\n--- Project Myrmidon: PheromoneTrail Module Test Complete ---")
```

```python
# Cell 3
# ==============================================================================
# TRINITY ðŸ‡: Project Myrmidon - QueenCore Module
# This module contains the "QueenCore," the central AI that analyzes the
# PheromoneTrail and orchestrates a collective response from the AntAgents.
# ==============================================================================
import json
from collections import Counter

class QueenCore:
    """
    The central intelligence of the Digital Ant Colony. It analyzes the
    PheromoneTrail to detect and respond to threats.
    """
    def __init__(self, pheromone_trail):
        self.pheromone_trail = pheromone_trail
        print("TRINITY: QueenCore initialized. Awaiting data from the PheromoneTrail.")

    def analyze_pheromones(self) -> dict:
        """
        Analyzes the pheromone trail to identify trends and threats.
        For this prototype, it simply counts the different types of threats.
        """
        trail_data = self.pheromone_trail.get_trail()
        if not trail_data:
            print("TRINITY: PheromoneTrail is empty. No analysis performed.")
            return {}

        threats = [p.get("threat_type") for p in trail_data]
        threat_counts = Counter(threats)

        print("\nTRINITY: QueenCore analysis complete.")
        print(f"TRINITY: Threat counts: {dict(threat_counts)}")
        return dict(threat_counts)

    def orchestrate_response(self, threat_counts: dict) -> None:
        """
        Orchestrates a simulated response based on the threat analysis.
        """
        if threat_counts.get("malicious_scan", 0) > 1:
            print("TRINITY: Escalated threat detected (multiple malicious scans). Initiating 'swarm lockdown' protocol.")
            # In a full system, this would trigger a series of actions across the network
        elif threat_counts.get("suspicious_event", 0) > 2:
            print("TRINITY: Potential coordinated attack detected. Initiating 'collective scrutiny' protocol.")
        else:
            print("TRINITY: No major collective threat detected. Maintaining 'patrolling' status.")

if __name__ == "__main__":
    # Test block to demonstrate the full, integrated system
    from ant_agent import AntAgent
    from pheromone_trail import PheromoneTrail

    print("--- Project Myrmidon: Integrated System Test Initiated ---")

    # 1. Initialize the system components
    pheromone_trail = PheromoneTrail()
    queen_core = QueenCore(pheromone_trail)

    agent_a = AntAgent(node_id="gateway_01")
    agent_b = AntAgent(node_id="web_server_03")
    agent_c = AntAgent(node_id="db_server_05")

    # 2. Simulate agent activity and populate the trail
    print("\n--- Simulating 10 cycles of agent activity ---")
    for _ in range(10):
        obs_a = agent_a._simulate_observation()
        pher_a = agent_a.generate_pheromone(obs_a)
        if pher_a:
            pheromone_trail.add_pheromone(pher_a)

        obs_b = agent_b._simulate_observation()
        pher_b = agent_b.generate_pheromone(obs_b)
        if pher_b:
            pheromone_trail.add_pheromone(pher_b)

        obs_c = agent_c._simulate_observation()
        pher_c = agent_c.generate_pheromone(obs_c)
        if pher_c:
            pheromone_trail.add_pheromone(pher_c)

    # 3. QueenCore analyzes the collected data and orchestrates a response
    threat_analysis = queen_core.analyze_pheromones()
    queen_core.orchestrate_response(threat_analysis)

    print("\n--- Project Myrmidon: Integrated System Test Complete ---")
```
